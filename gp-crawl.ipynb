```python
import requests
from bs4 import BeautifulSoup
import pymongo
import time
tdb_path = ''
tdb = pymongo.MongoClient(tdb_path)['model_train']
url_head = 'https://apkcombo.com/'
for i in range(1000):
#   headers = {
#     "accept": "text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3",
#     "cookie": "__cfduid=d9418c4b77ba3816995f153425677576f1571301759; _ga=GA1.2.642411827.1571301764; _gid=GA1.2.1404698229.1571301764; cf_clearance=b5e8c8c60e630f0925f481efb489bf5deb64c016-1571377591-0-150",
#     "referer": "https://apkcombo.com/zh-in/category/finance/latest-updates/?page=1",
#     "user-agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/77.0.3865.120 Safari/537.36"
#   }
  url = 'https://apkcombo.com/en-in/category/finance/latest-updates/?page=' + str(i)
  res = requests.get(url)
  bs = BeautifulSoup(res.text, 'html.parser')
  applist = bs.find('div', class_='columns is-multiline').find_all('a')
  for app in applist:
    app_name = app.find('b').text
    appurl = url_head + app['href']
    res = requests.get(appurl)
    bs = BeautifulSoup(res.text, 'html.parser')
    try:
      descript_1 = bs.find('div', style='margin-bottom: 20px').find('h2').text
    except:
      descript_1 = ''
    try:
      descript_more = bs.find('div', id='show-more-content').find('div').text
    except:
      descript_more = ''
    description = descript_1 + descript_more
    try:
      package = bs.find('table', class_='atable is-striped').find('a').text
    except:
      package = ''
    try:
      infolist = bs.find('table', class_='atable is-striped').find_all('td')
      version = infolist[3].text
      date = infolist[5].text
      size = infolist[7].text
      install_num = infolist[9].text
    except:
      version = version if version != '' else ''
      date = date if date != '' else ''
      size = size if size != '' else ''
      install_num = install_num if install_num != '' else ''
    data = dict()
    data['app_name'] = app_name
    data['description'] = description
    data['package'] = package
    data['version'] = version
    data['date'] = date
    data['size'] = size
    data['install_num'] = install_num
    tdb.app_list_spider.insert(data)
    print(app_name)
  time.sleep(5)

```
